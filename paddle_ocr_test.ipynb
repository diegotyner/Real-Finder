```python
def closest_partition(bounding_box, slices):
  max_x = 2400
  max_y = 400

  x_center = (bounding_box[0][0]+bounding_box[1][0]) / 2
  y_center = (bounding_box[0][1]+bounding_box[2][1]) / 2

  width_partition = round((x_center / max_x)*slices) / slices
  height_partition = round((y_center / max_y)*slices) / slices

  # print("Centers: ",x_center, y_center)
  # print("Closest partition: ", width_partition, height_partition)
  return [width_partition, height_partition]

def box_classifier(rec_text: str, confidence: float, bounding_box: list, page_number: int):
  max_x = 2400
  max_y = 400
  # classes - feel (genre?), Name, Artists, Page num, Other

  number_to_class_dict = {
    "feel": 0,
    "song_name": 1,
    "artist": 2,
    "page_num": 3,
    "other": 4,
    "empty": 5,
    "unknown": 6
  }

  if confidence == 0:
    return number_to_class_dict["empty"]
  
  try:
    rec_text_num = int(rec_text)
    if rec_text_num == page_number + 2:
      return number_to_class_dict["page_num"]
  except ValueError:
    pass
  
  width_center, height_center = closest_partition(bounding_box, 8)
  
  if rec_text.startswith("-") or ( (0.75 <= width_center <= 1) and (0.5 <= height_center <= 0.75) ):
    if confidence >= 0.5:
      return number_to_class_dict["artist"]
    else:
      print("artist, but no confidence")
  
  if (0.375 <= width_center <= 0.625) and (0.375 <= height_center <= 0.625):
    if confidence >= 0.8:
      return number_to_class_dict["song_name"]
    else:
      print("center box, but no confidence")

  if (rec_text.startswith("(") and rec_text.endswith(")")) or ((0 <= width_center <= 0.25) and (0.5 <= height_center <= 0.75)):
    return number_to_class_dict["feel"]
  
  return number_to_class_dict["unknown"]

  
  
  
  
```

```python
corner_bounds = [float('inf'), float('-inf'), float('inf'), float('-inf')]
for page_number in range(280, 290):
  output_folder_name = os.path.join("output", str(page_number), f"cropped_im_{page_number}_res.json")
  with open(output_folder_name) as f:
    data = json.load(f)
    
    detected_boxes = len(data.get("rec_texts", []))
    print(detected_boxes)
    for i in range(detected_boxes):
      print(f"\t Recognized: {data['rec_texts'][i]}")
      print(f"\t\t Score: {data['rec_scores'][i]}")
      print(f"\t\t Box: {data['rec_polys'][i]}")
      for j in range(4):
        # x check
        corner_bounds[0] = min(data.get("rec_polys")[i][j][0], corner_bounds[0])
        corner_bounds[1] = max(data.get("rec_polys")[i][j][0], corner_bounds[1])

        # y check
        corner_bounds[2] = min(data.get("rec_polys")[i][j][1], corner_bounds[2])
        corner_bounds[3] = max(data.get("rec_polys")[i][j][1], corner_bounds[3])

print(corner_bounds)  # Returned [47, 2399, 37, 400], which implies its around X - [0,2400] and Y - [0, 400] at 300 dpi
```

```python
number_to_class_dict = {
    0: "feel",
    1: "song_name",
    2: "artist",
    3: "page_num",
    4: "other",
    5: "empty",
    6: "unknown"
}

for page_number in range(230, 240):
  output_folder_name = os.path.join("output", str(page_number), f"cropped_im_{page_number}_res.json")
  with open(output_folder_name) as f:
    data = json.load(f)
    detected_boxes = len(data.get("rec_texts", []))
    print("\n- - - - - - - - - -\n")
    for i in range(detected_boxes):
      print(f"\t Recognized: {data['rec_texts'][i]}")
      class_pred = box_classifier(data['rec_texts'][i], data['rec_scores'][i], data['rec_polys'][i], page_number)
      print(f"\t\t Predicted: {number_to_class_dict[class_pred]}")
      # print(f"\t\t Score: {data['rec_scores'][i]}")
      # print(f"\t\t Box: {data['rec_polys'][i]}")
      # print(f"\t\t X/Y Partitions: {closest_partition(data['rec_polys'][i], 8)}")
```
